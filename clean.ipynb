```python
import re
```

```python
def clean_job_title(title):
    if pd.isna(title):
        return title
    
    s = str(title)
    
    # 1. Bỏ số + dấu phẩy ở đầu
    s = re.sub(r'^\d+,\s*', '', s)
    
    # 2. Bỏ tất cả nội dung trong () và []
    s = re.sub(r'\(.*?\)', '', s)
    s = re.sub(r'\[.*?\]', '', s)
    
    # 3. Bỏ các từ khóa tuyển dụng dư thừa
    s = re.sub(r'Tuyển Dụng|TUYỂN DỤNG', '', s, flags=re.IGNORECASE)
    
    # 4. Bỏ ký tự _ hoặc - dư
    s = re.sub(r'[_\-]+', ' ', s)
    
    # 5. Chuẩn hóa khoảng trắng
    s = re.sub(r'\s+', ' ', s)
    
    return s.strip()
```

```python
dfs = {
    "df_1" : df_1,
    "df_2" : df_2,
    "df_3" : df_3
}
for name, df in dfs.items():
    if 'job_title' in df.columns:
        df['job_title'] = df['job_title'].apply(clean_job_title)
        df.dropna(subset=['job_title'], inplace=True)
```

```python
def format_date(date):
    if pd.isna(date):
        return pd.NaT

    date = str(date)

    # Book3 – ISO 8601
    if 'T' in date:
        return pd.to_datetime(date, errors='coerce').date()

    # Book1, Book2 – dd/mm/yyyy
    return pd.to_datetime(date, format="%d/%m/%Y", errors='coerce').date()
```

```python
for name, df in dfs.items():
    if 'date' in df.columns:
        df['date'] = df['date'].apply(format_date)
        df.dropna(subset=['date'], inplace = True)
```

```python
import numpy as np
```

```python
def salary_shortcut(salary):
    if pd.isna(salary):
        return np.nan
    
    s = str(salary).lower()
    
    if any(x in s for x in ['thỏa thuận', 'thương lượng', 'cạnh tranh']):
        return np.nan
    
    nums = re.findall(r'\d+[.,]?\d*', s)
    nums = [float(n.replace(',', '')) for n in nums]
    
    if not nums:
        return np.nan
    
    max_value = max(nums)
    
    if 'triệu' in s or 'tr' in s:
        return max_value
    
    if 'usd' in s or '$' in s:
        return round(((max_value * 27000) / 1000000), 2)
    
    
    return np.nan
```

```python
for name, df in dfs.items():
    if 'salary' in df.columns:
        df['salary'] = df['salary'].apply(salary_shortcut)
        df.dropna(subset=['salary'], inplace = True)
       
```

```python
def location_clean(location):
    if pd.isna(location):
        return np.nan
    
    s = str(location)
    
    # 1. Bỏ nội dung trong ngoặc ()
    s = re.sub(r'\(.*?\)', '', s)
    
    # 2. Nếu có nhiều địa điểm, lấy cái đầu tiên
    s = s.split(',')[0]
    
    # 3. Chuẩn hóa khoảng trắng
    s = re.sub(r'\s+', ' ', s)
    
    return s.strip()
```

```python
for name, df in dfs.items():
    if 'location' in df.columns:
        df['location'] = df['location'].apply(location_clean)
        df.dropna(subset=['location'], inplace = True)
```
